```python
import matplotlib.pyplot as plt
import numpy as np

x  = np.arange(0, 10, 0.01)

bias = (x*x) / ((1+x)*(1+x))
variance = 9 / (10*(1+x)*(1+x))
decomposition = bias + variance

plt.title("Bias-Variance Decomposition")
plt.plot(x, bias)
plt.plot(x, variance)
plt.plot(x, decomposition)
plt.legend(["bias", "variance", "decomposition"], loc = 'upper left')

plt.show()
```
